```python
import matplotlib.pyplot as plt #importamos las bibliotecas
import numpy as np
import pandas as pd
```

```python
link="https://github.com/ChrisCT17/TareasTHC/blob/main/ECOVSCOVID.csv"#Link de github
code="?raw=true" #para que pandas lea el archivo csv y se pueda procesar
url=link+code #se unen el link y el codigo para leer el archivo
df=pd.read_csv(url,sep=",",index_col=1) #cargar archivo a la variable df
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Asegurarse de que la columna DATE esté en formato datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Rellenar los valores faltantes de HDI con el promedio
df['HDI'] = df['HDI'].fillna(df['HDI'].mean())

# Calcular el HDI promedio por fecha
hdi_over_time = df.groupby('DATE')['HDI'].mean()

# Graficar HDI respecto al tiempo
plt.figure(figsize=(12, 6))
plt.plot(hdi_over_time.index, hdi_over_time.values, label='HDI Promedio', color='blue')
plt.title('Evolución del HDI Promedio en el Tiempo', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('HDI Promedio', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Asegurarse de que la columna DATE esté en formato datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Calcular el total acumulado por fecha para TC y TD
tc_over_time = df.groupby('DATE')['TC'].sum()
td_over_time = df.groupby('DATE')['TD'].sum()

# Crear la gráfica combinada
plt.figure(figsize=(12, 6))
plt.plot(tc_over_time.index, tc_over_time.values, label='Casos Confirmados (TC)', color='red')
plt.plot(td_over_time.index, td_over_time.values, label='Muertes Totales (TD)', color='purple')
plt.title('Evolución de Casos Confirmados (TC) y Muertes Totales (TD) en el Tiempo', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Totales', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Asegurarse de que la columna DATE esté en formato datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Calcular el promedio de STI por fecha
sti_over_time = df.groupby('DATE')['STI'].mean()

# Graficar STI respecto al tiempo
plt.figure(figsize=(12, 6))
plt.plot(sti_over_time.index, sti_over_time.values, label='Rigor de políticas gubernamentales (STI)', color='green')
plt.title('Evolución de Rigor de políticas gubernamentales (STI) en el Tiempo', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('STI Promedio', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
