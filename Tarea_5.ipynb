##**2. Multiplicación entrada a entrada de multiplicaciones rectangulares**

```python
# Definir las matrices
A = [[11, 7, -1], [-3, 5, 2],   [0, 2, 3]]

B = [ [1, 1, 1],  [10, 3, -5],  [6, -10, -3]]

# crear una matriz vacia para poder almacenar el resultado
resultado = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]

# Multiplicar entrada por entrada
for i in range(len(A)):  # Recorrer las filas
    for j in range(len(A[0])):  # Recorrer las columnas
        resultado[i][j] = A[i][j] * B[i][j]

# Imprimir el resultado
print("Producto entrada por entrada:")
for fila in resultado:
    print(fila)
```

## **3. La traza de una matriz**

```python
# Definir una matriz
matriz = [[11, 7, -1], [-3, 5, 2], [0, 2, 3]]

# Inicializar la variable para la traza
traza = 0

# Calcular la traza (suma de los elementos diagonales)
for i in range(len(matriz)):
    traza += matriz[i][i]  # Sumar los elementos de la diagonal principal

# Imprimir el resultado
print("La traza de la matriz es:", traza)
```

## **5. Maximo y minimo de una matriz**

```python
matriz_2 = [ #Definimos las matrices
    [11, 2, 41],
    [9, 45, 14],
    [55,86, 99]
]

def maximo_y_minimo(matriz_2): #Definimos la funcion que encuentra el maximo y el minimo elemento de una matriz
  maximo=matriz_2[0][0] #Damos los valores iniciales del minimo y el maximo
  minimo=matriz_2[0][0]
  matriz_2=np.array(matriz_2) #Convvertimos a arreglo la matriz
  filas,columnas=matriz_2.shape #Obtenemos las dimensiones de la matriz

  for filas in matriz_2: #Recorremos la matriz
    for elemento in filas: #Reoprremos cada elemento de la matriz
      if elemento > maximo: #Dado que el maximo es mayor que el resto de elementos, buscamos que elemento es igual a este maximo
        maximo = elemento
      if elemento < minimo: #Dado que el minimo es menor que el resto de elementos, buscamos que elemento es igual a este minimo
        minimo = elemento
  print("El maximo elemento de la matriz es: ",maximo) #Imprimimos los minimos y los maximos
  print("El minimo elemento de la matriz es: ",minimo)
maximo_y_minimo(matriz_2)
```

##**7. La potencia de matrices**

```python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
exponente = 2

def potencia_matriz(matriz, exponente):


    matriz = np.array(matriz)  # Convertir a un array de NumPy
    filas, columnas = matriz.shape

    if filas != columnas:
        print("Error: La matriz no es cuadrada.")
        return None

    if exponente < 0:
        print("Error: El exponente no puede ser negativo.")
        return None

    # Si el exponente es 0, devolvemos la matriz identidad
    if exponente == 0:
        return [[1 if i == j else 0 for j in range(filas)] for i in range(filas)]

    # Iniciamos la matriz resultado como la matriz original
    resultado = matriz.copy()

    for _ in range(exponente - 1):
        resultado = np.dot(resultado, matriz)

    return resultado

resultado = potencia_matriz(matriz, exponente)
print(resultado)
```

## **1. Solucion general**

```python
def solucion_general(x,c):
  return (x - 1) + c * np.exp(-x) #Definimos la funcion de la solucion genera, utuilizamos np.exp(-x) para la exponencial

c_positivo=np.linspace(1,10,10) #Damos los valores de c en cada caso, cuando es positivo, negativo y cuando vale 0
c_negativo=np.linspace(-10,-1,10)
c_cero=[0]
x=np.linspace(-2,4,100) #Damos los valores de x

#Creamos una grafica para cada caso
for c in c_positivo: #Graficamos C positivo
  if c==1: #Damos la primera curva con etiqueta para la leyenda
    plt.plot(x,solucion_general(x,c),label="C>0",color="blue")
  else:
    plt.plot(x,solucion_general(x,c),color="blue") #Graficamos el resto de curvas

for c in c_negativo: #Graficamos C negativo
  if c==-1: #Damos la primera curva con etiqueta para la leyenda
    plt.plot(x,solucion_general(x,c),label="C<0",color="purple")
  else:
    plt.plot(x,solucion_general(x,c),color="purple") #Graficamos el resto de curvas

for c in c_cero: #Graficamos C cuando vale 0
  plt.plot(x,solucion_general(x,c),label=f"C={c}",color="red",linestyle="--")

plt.xlabel("x",fontsize=12)
plt.ylabel("y",fontsize=12)
plt.title("Solución General de la Ecuación Diferencial",color="black",fontsize=14)
plt.legend(title=("Valores de C"))
plt.grid(True)

plt.show()
```
